```python
import numpy as np
import matplotlib.pyplot as plt
from sys import getsizeof
```

# <b> 1. One vector </b>

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.set_aspect('equal')

# Data
x_pos = 1
y_pos = 1
x_dir = 4
y_dir = 7

# (X, Y, X_dir, Y_dir)
ax.quiver(x_pos, y_pos, x_dir, y_dir,
          angles='xy', scale_units='xy', scale=1,
          width=0.01,       # szerokość pręta
          headwidth=5,
          headaxislength=5, # wcięcie
          headlength=5,
          color='black')

# Plot settings
xlims = [-2, 10]
ylims = [-2, 10]

ax.hlines(y=0, xmin=xlims[0], xmax=xlims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)
ax.vlines(x=0, ymin=ylims[0], ymax=ylims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)

ax.set_xlim(xlims)
ax.set_ylim(ylims)
ax.xaxis.set_ticks([i for i in range(xlims[0], xlims[1]+1)])
ax.set_yticks([i for i in range(ylims[0], ylims[1]+1)])
ax.grid()

ax.set_title('wektor')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
```

# <b> 2. Many vectors </b>

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Data
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)

X, Y = np.meshgrid(x, y)

u = X/5
v = -Y/5

#
#   X & Y are 2D arrays and contain vectors starting positions
#   u & v are 2D arrays and contain vectors end position
#
ax.quiver(X, Y, u, v,
          angles='xy', scale_units='xy',
          scale=2,
          width=0.0035,       # szerokość pręta
          headwidth=5,
          headaxislength=5, # wcięcie
          headlength=5,
          color='blue')

# Plot settings

xlims = [-1, 1]
ylims = [-1, 1]

ax.hlines(y=0, xmin=xlims[0], xmax=xlims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)
ax.vlines(x=0, ymin=ylims[0], ymax=ylims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)

ax.set_xlim(xlims)
ax.set_ylim(ylims)
xticks = [i for i in np.arange(xlims[0], xlims[1] + 0.1, 0.1)]
yticks = [i for i in np.arange(ylims[0], ylims[1] + 0.1, 0.1)]
ax.xaxis.set_ticks(xticks)
ax.yaxis.set_ticks(yticks)
ax.set_aspect('equal')
ax.grid()

ax.set_title('wektor')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
```

# <b> Normalized vectors </b>

```python
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)

X, Y = np.meshgrid(x, y)

u = X/5
v = -Y/5

# normalize vectors
vec_len = np.sqrt( np.power(u, 2) + np.power(v, 2) )
scale = 0.15
u = u/vec_len * scale
v = v/vec_len * scale

fig, ax = plt.subplots(figsize=(8, 8))
ax.quiver(X, Y, u, v,
          angles='xy', scale_units='xy',
          scale=2,
          width=0.0035,       # szerokość pręta
          headwidth=5,
          headaxislength=5, # wcięcie
          headlength=5,
          color='blue')
```

# <b> 3. Gradient of a functions </b>

```python
#
#   Vector field
#
x = np.arange(-1, 1, 0.075)
y = np.arange(-1, 1, 0.075)

X, Y = np.meshgrid(x, y)
z = x*np.exp(-X**2 -Y**2)
v, u = np.gradient(f=z)

# Normalization
vec_len = np.sqrt( np.power(u, 2) + np.power(v, 2) )
scale = 0.03
u = u/vec_len * scale
v = v/vec_len * scale

# Figure
fig = plt.figure(figsize=(15, 6))
fig.suptitle('Super plots')

ax = fig.add_subplot(1, 1, 1)
#
#   X & Y are 2D arrays and contain vectors starting positions
#   u & v are 2D arrays and contain vectors end position
#
ax.quiver(X, Y, u, v,
          angles='xy', scale_units='xy',
          scale=0.5,
          width=0.0035,       # szerokość pręta
          headwidth=5,
          headaxislength=5, # wcięcie
          headlength=5,
          color='blue')

# Contour plot
ax.contour(X, Y, z, levels=15, linewidths=3)

# Plot settings
xlims = [-1, 1]
ylims = [-1, 1]

ax.hlines(y=0, xmin=xlims[0], xmax=xlims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)
ax.vlines(x=0, ymin=ylims[0], ymax=ylims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)

ax.set_xlim(xlims)
ax.set_ylim(ylims)
xticks = [i for i in np.arange(xlims[0], xlims[1] + 0.1, 0.1)]
yticks = [i for i in np.arange(ylims[0], ylims[1] + 0.1, 0.1)]
ax.xaxis.set_ticks(xticks); ax.set_xticklabels(np.round(ax.get_xticks(), 2), rotation=55)
ax.yaxis.set_ticks(yticks)
ax.set_aspect('equal')
ax.grid()

ax.set_title('wektor')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# #
# #   3D plot
# #
# ax = fig.add_subplot(1, 2, 2, projection='3d')
# surf = ax.plot_surface(X, Y, z, rstride=1, cstride=1, linewidth=1, antialiased=False, shade=True)
# ax.set_xlabel('x axis')
# ax.set_ylabel('y axis')
# ax.set_zlabel('z axis')
```

### <b> 6. Stream plot </b>

```python
#
#   Vector field
#
x = np.arange(-1, 1, 0.05)
y = np.arange(-1, 1, 0.05)

X, Y = np.meshgrid(x, y)
z = X*np.exp(-X**2 -Y**2)
v, u = np.gradient(f=z)

# Normalization
vec_len = np.sqrt( np.power(u, 2) + np.power(v, 2) )
scale = 0.032
u = u/vec_len * scale
v = v/vec_len * scale

# Figure
fig = plt.figure(figsize=(15, 6))
fig.suptitle('Super plots')

ax = fig.add_subplot(1, 1, 1)
#
#   X & Y are 2D arrays and contain vectors starting positions
#   u & v are 2D arrays and contain vectors end position
#
ax.quiver(X, Y, u, v,
          angles='xy', scale_units='xy',
          scale=0.5,
          width=0.0035,       # szerokość pręta
          headwidth=5,
          headaxislength=5, # wcięcie
          headlength=5,
          color='blue',
          alpha=0.3,
          zorder=2)

#
#   Streamplot
#
ax.streamplot(X, Y, u, v,
              density=0.5, 
              color='red',
              linewidth=2,
              arrowsize=2,
              zorder=2,
              maxlength=4,
              minlength=0.1,
              broken_streamlines=True)

# # Contour plot
# # stram plot pokazywał by chyba jak iśc najbardziej pod górke w danym miejscu XD
# ax.contour(X, Y, z, levels=15, linewidths=3)

# Plot settings
xlims = [-1, 1]
ylims = [-1, 1]

ax.hlines(y=0, xmin=xlims[0], xmax=xlims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)
ax.vlines(x=0, ymin=ylims[0], ymax=ylims[1], colors='black', linestyles='-', linewidth=1, alpha=0.9)

ax.set_xlim(xlims)
ax.set_ylim(ylims)
xticks = [i for i in np.arange(xlims[0], xlims[1] + 0.1, 0.1)]
yticks = [i for i in np.arange(ylims[0], ylims[1] + 0.1, 0.1)]
ax.xaxis.set_ticks(xticks); ax.set_xticklabels(np.round(ax.get_xticks(), 2), rotation=55)
ax.yaxis.set_ticks(yticks)
ax.set_aspect('equal')
ax.grid()
```

### <b> 7. 3D vector field </b>

```python
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
z = np.arange(-1, 1, 0.1)

X, Y, Z = np.meshgrid(x, y, z)
k = X*np.exp(-np.power(X, 2) -np.power(Y, 2) -np.power(Z, 2))

# grad_vec = <u, v, s>
s, v, u = np.gradient(f=k)  # 3D vector field

# Normalization
vec_len = np.sqrt( u**2 + v**2 + s**2 )
scale = 0.08
u = u/vec_len * scale
v = v/vec_len * scale
s = s/vec_len * scale

# Figure
fig = plt.figure(figsize=(15, 15))

n_rows = 3
n_cols = 2
for i in range(1, n_rows*n_cols+1):
    ax = fig.add_subplot(n_rows, n_cols, i)

    ax.quiver(X[:, :, i], Y[:, :, i], u[:, :, i], v[:, :, i],
            angles='xy', scale_units='xy',
            scale=1,
            width=0.0035,       # szerokość pręta
            headwidth=5,
            headaxislength=5, # wcięcie
            headlength=5,
            color='blue',
            alpha=1,
            zorder=2)

```
